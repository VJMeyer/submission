### Build data frame with extractable information from BayesDown

```python
def parse_markdown_hierarchy(markdown_text):
    # Initialize an empty DataFrame with all required columns
    df = pd.DataFrame(columns=[
        'Title', 'Description', 'Parents', 'Children', 'No_Parent', 'No_Children'
    ])
    
    # Split the markdown text into lines
    lines = markdown_text.split('\n')
    
    # Regex pattern to extract text between angle brackets
    title_pattern = r'<(.*?)>'
    
    # Keep track of the last child title and its potential parents
    last_child_title = None
    current_parents = []
    
    # Process each line
    for line in lines:
        # Trim the line
        stripped_line = line.strip()
        
        # Find titles in the current line
        titles = re.findall(title_pattern, line)
        
        # Check if this is a main title line (not starting with "  + ")
        # Ensure it starts with a title enclosed in < > and doesn't start with "+ <"
        if titles and re.match(r'^<.*?>:', stripped_line) and not stripped_line.startswith('+ <'):
            for title in titles:
                # Extract description (text after ":")
                description_match = re.search(f'<{title}>: (.*)', line)
                description = description_match.group(1) if description_match else ''
                
                # Create a new row in the DataFrame
                new_row = {
                    'Title': title,
                    'Description': description,
                    'Parents': current_parents.copy(),  # Use a copy of current parents
                    'Children': [],
                    'No_Parent': len(current_parents) == 0,
                    'No_Children': True
                }
                
                # Add the row to the DataFrame
                df.loc[len(df)] = new_row
                
                # Update last child title
                last_child_title = title
                
                # Reset parents for the next main title
                current_parents = []
        
        # Check if this is a parent line (starting with "  + ")
        elif titles and stripped_line.startswith('+ <'):
            for title in titles:
                # Extract description (text after ":")
                description_match = re.search(f'\\+ <{title}>: (.*)', line)
                description = description_match.group(1) if description_match else ''
                
                # Add this title to current parents
                current_parents.append(title)
                
                # Create a new row in the DataFrame
                new_row = {
                    'Title': title,
                    'Description': description,
                    'Parents': [],  # No parent for this title
                    'Children': [last_child_title] if last_child_title else [],
                    'No_Parent': True,
                    'No_Children': last_child_title is None
                }
                
                # Add the row to the DataFrame
                df.loc[len(df)] = new_row
                
                # If there was a last child, update its parents
                if last_child_title:
                    # Find the index of the last child row
                    last_child_index = df[df['Title'] == last_child_title].index[-1]
                    df.at[last_child_index, 'Parents'] = current_parents.copy()
                    df.at[last_child_index, 'No_Parent'] = False
    
    return df


# Example usage
#markdown_text = """ # Main Argument: Current AI as Existential Risk Factor\n\n<AI_Existential_Risk_Factor>: Current and near-term AI technologies can contribute to existential risk by acting as intermediate risk factors.\n  + <AI_Intermediate_Factor>: Current and near-term AI technologies can act as intermediate risk factors, magnifying the likelihood of previously identified sources of existential risk.\n  + <Risk_Not_Limited_To_AGI>: This potential contribution to existential risk is not limited to the unaligned AGI scenario.\n  + <Causal_Pathways_Exist>: There exist causal pathways from AI systems to existential risks that do not presuppose hypothetical future AI capabilities.\n\n\n# Power Dynamics Arguments"""
markdown_text = md_content

# Parse the markdown
result_df = parse_markdown_hierarchy(markdown_text)

# Display the DataFrame
result_df
```

```python
def parse_markdown_hierarchy(markdown_text):
    # Remove comments first
    markdown_text = re.sub(r'/\*.*?\*/', '', markdown_text, flags=re.DOTALL)
    
    # Initialize an empty DataFrame with all required columns
    df = pd.DataFrame(columns=[
        'Title', 'Description', 'Parents', 'Children', 'No_Parent', 'No_Children'
    ])
    
    # Split the markdown text into lines
    lines = markdown_text.split('\n')
    
    # Regex pattern to extract text between angle brackets or square brackets
    title_pattern = r'[<\[](.+?)[>\]]'
    
    # Keep track of the last child title and its potential parents
    last_child_title = None
    current_parents = []
    nested_parents = []
    
    # Process each line
    for line in lines:
        # Trim the line
        stripped_line = line.strip()
        
        # Skip empty lines
        if not stripped_line:
            continue
        
        # Find titles in the current line
        titles = re.findall(title_pattern, line)
        
        # Check for nested parent levels
        if line.startswith('  +  '):
            # Two spaces after '+' indicates a nested parent
            nested_parent_match = re.findall(title_pattern, line)
            if nested_parent_match:
                nested_parents.append(nested_parent_match[0])
        
        # Check if this is a main title line (not starting with "  + ")
        # Ensure it starts with a title enclosed in < > or [ ] and doesn't start with "+ <" or "+ ["
        if (titles and 
            re.match(r'^[<\[].*?[>\]]:', stripped_line) and 
            not stripped_line.startswith('+ <') and 
            not stripped_line.startswith('+ [')):
            for title in titles:
                # Extract description (text after ":")
                description_match = re.search(fr'[<\[]{re.escape(title)}[>\]]: (.*)', line)
                description = description_match.group(1) if description_match else ''
                
                # Create a new row in the DataFrame
                new_row = {
                    'Title': title,
                    'Description': description,
                    'Parents': current_parents.copy(),  # Use a copy of current parents
                    'Children': [],
                    'No_Parent': len(current_parents) == 0,
                    'No_Children': True
                }
                
                # Add the row to the DataFrame
                df.loc[len(df)] = new_row
                
                # Update last child title
                last_child_title = title
                
                # Reset parents for the next main title
                current_parents = []
                nested_parents = []
        
        # Check if this is a parent line (starting with "  + ")
        elif titles and (stripped_line.startswith('+ <') or stripped_line.startswith('+ [') or stripped_line.startswith('  + ')):
            for title in titles:
                # Extract description (text after ":")
                description_match = re.search(fr'\+ [<\[]{re.escape(title)}[>\]]: (.*)', line)
                description = description_match.group(1) if description_match else ''
                
                # Add this title to current parents
                current_parents.append(title)
                
                # Create a new row in the DataFrame
                new_row = {
                    'Title': title,
                    'Description': description,
                    'Parents': nested_parents.copy() if nested_parents else [],  # Use nested parents if available
                    'Children': [last_child_title] if last_child_title else [],
                    'No_Parent': len(nested_parents) == 0,
                    'No_Children': last_child_title is None
                }
                
                # Add the row to the DataFrame
                df.loc[len(df)] = new_row
                
                # If there was a last child, update its parents
                if last_child_title:
                    # Find the index of the last child row
                    last_child_index = df[df['Title'] == last_child_title].index[-1]
                    
                    # Update parents to include both current and nested parents
                    combined_parents = current_parents.copy()
                    if nested_parents:
                        combined_parents.extend(nested_parents)
                    
                    df.at[last_child_index, 'Parents'] = combined_parents
                    df.at[last_child_index, 'No_Parent'] = len(combined_parents) == 0
    
    return df


# Example usage
markdown_text = md_content

# Parse the markdown
result_df = parse_markdown_hierarchy(markdown_text)

# Display the DataFrame
result_df
```

```python
# implement with additional argument generation (i.e. until "grandparents")

def parse_markdown_hierarchy(markdown_text):
    # Remove comments first
    markdown_text = re.sub(r'/\*.*?\*/', '', markdown_text, flags=re.DOTALL)
    
    # Initialize an empty DataFrame with all required columns
    df = pd.DataFrame(columns=[
        'Title', 'Description', 'Parents', 'Children', 'No_Parent', 'No_Children'
    ])
    
    # Split the markdown text into lines
    lines = markdown_text.split('\n')
    
    # Regex pattern to extract text between angle brackets or square brackets
    title_pattern = r'[<\[](.+?)[>\]]'
    
    # Keep track of hierarchy
    hierarchy = []
    last_title = None
    
    # Process each line
    for line in lines:
        # Trim the line
        stripped_line = line.strip()
        
        # Skip empty lines
        if not stripped_line:
            continue
        
        # Find titles in the current line
        titles = re.findall(title_pattern, line)
        
        # Determine the indentation level (number of spaces before '+')
        indent_match = re.match(r'^(\s*)\+ ', stripped_line)
        indent_level = len(indent_match.group(1)) if indent_match else 0
        
        # Adjust hierarchy based on indentation
        while len(hierarchy) > indent_level:
            hierarchy.pop()
        
        # Process main titles (direct headers)
        if (titles and 
            re.match(r'^[<\[].*?[>\]]:', stripped_line) and 
            not stripped_line.startswith('+ <') and 
            not stripped_line.startswith('+ [')):
            for title in titles:
                # Extract description (text after ":")
                description_match = re.search(fr'[<\[]{re.escape(title)}[>\]]: (.*)', line)
                description = description_match.group(1) if description_match else ''
                
                # Create a new row in the DataFrame
                new_row = {
                    'Title': title,
                    'Description': description,
                    'Parents': hierarchy.copy(),
                    'Children': [],
                    'No_Parent': len(hierarchy) == 0,
                    'No_Children': True
                }
                
                # Add the row to the DataFrame
                df.loc[len(df)] = new_row
                
                # Update last title and hierarchy
                last_title = title
                if indent_level + 1 == len(hierarchy):
                    hierarchy.append(title)
                elif indent_level == len(hierarchy):
                    hierarchy[-1] = title
                
        # Process parent/child lines (starting with "+ " or "  + ")
        elif titles and (stripped_line.startswith('+ <') or 
                         stripped_line.startswith('+ [') or 
                         stripped_line.startswith('  + ')):
            for title in titles:
                # Extract description (text after ":")
                description_match = re.search(fr'\+ [<\[]{re.escape(title)}[>\]]: (.*)', line)
                description = description_match.group(1) if description_match else ''
                
                # Create a new row in the DataFrame
                new_row = {
                    'Title': title,
                    'Description': description,
                    'Parents': hierarchy.copy(),
                    'Children': [last_title] if last_title else [],
                    'No_Parent': len(hierarchy) == 0,
                    'No_Children': last_title is None
                }
                
                # Add the row to the DataFrame
                df.loc[len(df)] = new_row
                
                # Update hierarchy and last title
                if indent_level + 1 == len(hierarchy):
                    hierarchy.append(title)
                elif indent_level == len(hierarchy):
                    hierarchy[-1] = title
                
                # Update previous title's children if applicable
                if last_title:
                    # Find the index of the last title row
                    last_title_index = df[df['Title'] == last_title].index[-1]
                    
                    # Update children
                    children = df.at[last_title_index, 'Children']
                    if title not in children:
                        children.append(title)
                    df.at[last_title_index, 'Children'] = children
                    df.at[last_title_index, 'No_Children'] = False
                
                # Update last title
                last_title = title
    
    return df


# Example usage
markdown_text = md_content

# Parse the markdown
result_df = parse_markdown_hierarchy(markdown_text)

# Display the DataFrame
result_df

```
